```python
#Importing all necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Loading the dataset
df1=pd.read_csv("/content/Unemployment in India.csv")
df2=pd.read_csv("/content/Unemployment_Rate_upto_11_2020.csv")

#display the data
print(df1)
print(df2)
```

```python
#to check if datasets contain null values
df1.isnull().sum()
df2.isnull().sum()
```

```python
# drop the null values
df1.dropna()
df2.dropna()
```

```python
#Standardize the data of df1 and df2
df1.columns=df1.columns.str.strip().str.lower().str.replace(' ', '_')
df2.columns=df2.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df1['date']=pd.to_datetime(df1['date'].str.strip(),format='%d-%m-%Y')
df2['date']=pd.to_datetime(df2['date'].str.strip(),format='%d-%m-%Y')
```

```python
#Performing Exploratory Data Analysis
#1. Analyzing overall trends
import seaborn as sns
sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
sns.lineplot(x='date', y='estimated_unemployment_rate_(%)', data=df1)
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.show()
plt
```

```python
# Showing top 5 refions with high unemployment rates
region_unemployment = df1.groupby('region')['estimated_unemployment_rate_(%)'].mean().sort_values(ascending=False).head(5)
top_5_regions=region_unemployment.head(5
                                       )
plt.figure(figsize=(10,6))
plt.bar(top_5_regions.index, top_5_regions.values)
plt.title('Top 5 Regions with High Unemployment Rates')
plt.xlabel('Region')
plt.ylabel('Average Unemployment Rate')
plt.show()
```

```python
# Rural vs Urban Trends
plt.figure(figsize=(10,6))
sns.lineplot(data=df1, x='date', y='estimated_unemployment_rate_(%)', hue='area')
plt.title('Rural vs Urban Unemployment Trends')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.legend(title='Area', loc='upper right')
plt.show()
```

```python
# Heatmap of unemployment rates by region and Date
pivot_df = df1.pivot_table(index='date', columns='region', values='estimated_unemployment_rate_(%)', aggfunc='mean')
plt.figure(figsize=(15,10))
sns.heatmap(pivot_df, cmap='YlGnBu')
plt.title('Heatmap of Unemployment Rates by Region and Date')
plt.show()
```

```python
#common regions during covid-19
covid_period_start='2020-03-01'
covid_period_end='2020-06-30'
# Filter the DataFrame for the COVID period
covid_df = df1[(df1['date'] >= covid_period_start) & (df1['date'] <= covid_period_end)]
# Get unique regions within the filtered DataFrame
covid_regions = covid_df['region'].unique()
# Filter the original DataFrame to include only regions present during the COVID period
coivid_regions = df1[df1['region'].isin(covid_regions)]
# Group by region and calculate the mean unemployment rate, then sort and get the top 5
top_5_regions_covid = coivid_regions.groupby('region')['estimated_unemployment_rate_(%)'].mean().sort_values(ascending=False).head(5)
plt.figure(figsize=(10,6))
plt.bar(top_5_regions_covid.index, top_5_regions_covid.values)
plt.title('Top 5 Regions with High Unemployment Rates During Covid-19')
plt.xlabel('Region')
plt.ylabel('Average Unemployment Rate')
plt.show()
```

```python
#common top regions during Covid 19
covid_period_start='2020-03-01'
covid_period_end='2020-06-30'
covid_data1=df1[(df1['date']>=covid_period_start) & (df1['date']<=covid_period_end)]
covid_data2=df2[(df2['date']>=covid_period_start) & (df2['date']<=covid_period_end)]
top_regions1=covid_data1.groupby('region')['estimated_unemployment_rate_(%)'].mean().sort_values(ascending=False).head(5)
top_regions2=covid_data2.groupby('region')['estimated_unemployment_rate_(%)'].mean().sort_values(ascending=False).head(5)
common_regions=top_regions1[top_regions1.index.isin(top_regions2.index)]
plt.figure(figsize=(10,6))
plt.bar(common_regions.index, common_regions.values)
plt.title('Common Regions with High Unemployment Rates During Covid-19')
plt.xlabel('Region')
plt.ylabel('Average Unemployment Rate')
```
